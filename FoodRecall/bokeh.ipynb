```python
## Interactive visualization (Hover) by using Bokeh
## Boxplot of Durations (months) of Three Main Problem Types
import pandas as pd
from bokeh.charts import BoxPlot, output_file, show
from bokeh.io import output_notebook

df = pd.read_csv("DurationAnalysis.csv",encoding='latin1')
df.head()
```

```python
#output_file("boxplot.html")
output_notebook()
```

```python
tooltips1=[
    ('Problem Type', '@problem_type'),
    ('Year', '@year')]
p = BoxPlot(df, values='duration_groups(month)', label=['problem_type','year'], color='problem_type',
            legend=False , title="Boxplot of Durations (months) of Three Main Problem Types", tooltips=tooltips1)
show(p)
```
